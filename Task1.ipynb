```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic_data = pd.read_csv('/content/sample_data/titanic.csv')

# Create a bar chart for gender distribution
gender_counts = titanic_data['Sex'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(gender_counts.index, gender_counts.values, color=['skyblue', 'lightcoral'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution on the Titanic')
plt.show()
```

```python
# Create a histogram for age distribution
plt.figure(figsize=(10, 6))
plt.hist(titanic_data['Age'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution on the Titanic')
plt.show()
```
